```python
# 3. Data Cleaning
import numpy as np

# to print the dataframe for refrence
df
```

```python
# Fill missing values with mean  if any column has missing values but you can see there is no value missing but we can fill by mean as below

df.fillna(df.mean(), inplace=True)
```

```python
# Drop irrelevant columns (example: 'Id') as it has same number as S.No so we can remove there will no impact 
df = df.drop(['Id'], axis=1, inplace=True)
```

```python
# Data Analysis
# Distribution of Cover_Type
print(df['Cover_Type'].value_counts())
```

```python
# Compare average elevation by cover type
print(df.groupby('Cover_Type')['Elevation'].mean())
```

```python
# Trend by slope and cover type
print(df.groupby('Cover_Type')['Slope'].mean())
```

```python
import matplotlib.pyplot as plot

# Visualizations
#  Horizontal vs Vertical Distance to Hydrology (Sum)
df.groupby("Horizontal_Distance_To_Hydrology")["Vertical_Distance_To_Hydrology"].sum().plot.bar(
    title="Sum of Vertical Distance for Each Horizontal Distance", figsize=(12, 5)
)
```

```python

# Pie Chart for Cover Type Proportion

df['Cover_Type'].value_counts().sort_index().plot.pie(
    autopct='%1.1f%%',
    figsize=(8, 8),
    title='Cover Type Proportion',
 
)
plot.show()

```
